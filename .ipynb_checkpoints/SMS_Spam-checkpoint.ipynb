```python
import pandas as pd
```

```python
data = pd.read_csv("email.csv")
data
```

```python
data.isnull().sum()
```

```python
X = data.drop(columns=["Category"])
X.head(10)
```

```python
y = data["Category"]
y.head(10)
```

```python
from sklearn.model_selection import train_test_split

# Step 1: Split the dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Step 2: Initialize the TF-IDF Vectorizer
vectorizer = TfidfVectorizer()

# Step 3: Flatten the data if necessary (only if X is a DataFrame)
X_train_flat = X_train.values.ravel()
X_test_flat = X_test.values.ravel()

# Step 4: Fit the vectorizer on the training data and transform both sets
X_train_tr = vectorizer.fit_transform(X_train_flat).toarray()  
X_test_tr = vectorizer.transform(X_test_flat).toarray()        

# Step 5: Check the shapes
print(X_train_tr.shape)  
print(y_train.shape)     
```

```python
from sklearn.naive_bayes import GaussianNB

# Step 6: Train the Gaussian Naive Bayes model
model = GaussianNB()
model.fit(X_train_tr, y_train)
```

```python
# Step 7: Make predictions on the test set
prediction = model.predict(X_test_tr)
print("Prediction", "Actual")
for i in range(len(prediction)):
    print(prediction[i], y_test.iloc[i])
```

```python
from sklearn.metrics import accuracy_score
score = accuracy_score(y_test, prediction)
score
```

```python
pre = model.predict(vectorizer.transform(["Congratulations! You've won a $1,000 Walmart gift card. "]).toarray())
print(pre)
```
